```python
#FIND NEW PRODUCTS
dataset_id = "english-prescribing-data-epd" # Dateset ID
date_from = "201901" # Can be either "YYYYMM" or "earliest", default="earliest"
date_to = "202112" # Can be either "YYYYMM" or "latest", default="latest"
sql = (
    "SELECT YEAR_MONTH, REGIONAL_OFFICE_CODE, BNF_CODE, BNF_DESCRIPTION, SUM(ITEMS) AS ITEMS, SUM(TOTAL_QUANTITY) AS QUANTITY, SUM(ACTUAL_COST) AS COST "
    "{FROM_TABLE} "
    "WHERE (BNF_CODE LIKE '010301%' OR BNF_CODE LIKE '010305%') "
    "GROUP BY YEAR_MONTH, REGIONAL_OFFICE_CODE, BNF_CODE, BNF_DESCRIPTION"
)

# Fetch data using BSA API
df=bsa_utils.fetch_data(resource=dataset_id, date_from=date_from, date_to=date_to, sql=sql)

# Save data to file
df.to_csv("../data/data.csv", index=False)

display(df)
```
